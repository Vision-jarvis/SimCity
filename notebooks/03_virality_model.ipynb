# 03 — Virality Forecasting

Multi-horizon engagement forecast from an event history using the analytical Hawkes-SEIR forecaster.

```python
import numpy as np
from ml.forecasting.virality_forecaster import ViralityForecaster

# Synthetic burst: rapid early engagement that decays.
t0 = 1_700_000_000
history = [{'timestamp': t0 + i*1800, 'engagement': float(50*np.exp(-0.03*i)+np.random.rand()*5)} for i in range(40)]

fc = ViralityForecaster()
results = fc.forecast_multi_horizon(history, topic='demo-meme')
for h, r in results.items():
    print(f'{h}h horizon: virality={r.virality_score:.3f} cascade_p={r.cascade_probability:.3f} peak@{r.peak_time_hours}h')
```

## Plot the 24h forecast

```python
import matplotlib.pyplot as plt
r = results[24]
ts = [p['timestamp'] for p in r.predictions]
plt.plot(ts, [p['predicted_engagement'] for p in r.predictions], label='predicted')
plt.fill_between(ts, [p['confidence_lower'] for p in r.predictions], [p['confidence_upper'] for p in r.predictions], alpha=0.2)
plt.legend(); plt.title('24h engagement forecast'); plt.show()
```
